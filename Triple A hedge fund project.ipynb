```python
# import the packages to be used in the coding project
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.formula.api as smf
```

```python
# import the data frames for the stock data
fb = pd.read_csv("facebook.csv")
```

```python
ms = pd.read_csv("microsoft.csv")
```

```python
app = pd.read_csv("apple.csv")
```

```python
tsla = pd.read_csv("tsla.csv")
```

```python
fb.index = fb["Date"]
ms.index = ms["Date"]
app.index = app["Date"]
tsla.index = tsla["Date"]
del app["Date"]
del tsla["Date"]
del fb["Date"]
del ms["Date"]
```

```python
del tsla["Symbol"]
```

```python
# create a new column for the price of tommorow

fb["price1"] = fb["Close"].shift(-1)

ms["price1"] = ms["Close"].shift(-1)

app["price1"] = app["Close"].shift(-1)

tsla["price1"] = tsla["Close"].shift(-1)
```

```python
# To create a column for the price difference of the price of tommorow and today

fb["pricediff"] = fb["price1"] - fb["Close"]
ms["pricediff"] = ms["price1"] - ms["Close"]
app["pricediff"] = app["price1"] - app["Close"]
tsla["pricediff"] = tsla["price1"] - tsla["Close"]
```

```python
# To create a column for the price difference of the price of tommorow and today

fb["return"] = fb["pricediff"]/fb["Close"]
ms["return"] = ms["pricediff"]/ms["Close"]
app["return"] = app["pricediff"]/app["Close"]
tsla["return"] = tsla["pricediff"]/tsla["Close"]
```

```python
# To create a column for the direction basing on the price difference 
fb["direction"] = [1 if fb.loc[item,"pricediff"]>0 else -1 for item in fb.index]
ms["direction"] = [1 if ms.loc[item,"pricediff"]>0 else -1 for item in ms.index]
app["direction"] = [1 if app.loc[item,"pricediff"]>0 else -1 for item in app.index]
tsla["direction"] = [1 if tsla.loc[item,"pricediff"]>0 else -1 for item in tsla.index]
```

```python
# To create a column for the moving average
fb["ma10"] = fb["Close"].rolling(10).mean()
fb["ma50"] = fb["Close"].rolling(50).mean()
ms["ma10"] = ms["Close"].rolling(10).mean()
ms["ma50"] = ms["Close"].rolling(50).mean()
app["ma10"] = app["Close"].rolling(10).mean()
app["ma50"] = app["Close"].rolling(50).mean()
tsla["ma10"] = tsla["Close"].rolling(10).mean()
tsla["ma50"] = tsla["Close"].rolling(50).mean()
```

```python
# To Create a column to show the stocks bought using this strategy
fb["stock"] = [1 if fb.loc[item,"ma10"] > fb.loc[item,"ma50"] else 0 for item in fb.index]
ms["stock"] = [1 if ms.loc[item,"ma10"] > ms.loc[item,"ma50"] else 0 for item in ms.index]
app["stock"] = [1 if app.loc[item,"ma10"] > app.loc[item,"ma50"] else 0 for item in app.index]
tsla["stock"] = [1 if tsla.loc[item,"ma10"] > tsla.loc[item,"ma50"] else 0 for item in tsla.index]
```

```python
# Calculate the profit that is gained in each deal for when the stock is 1

fb["profit"] = [fb.loc[item,"price1"] - fb.loc[item,"Close"] if fb.loc[item,"stock"]==1 else 0 for item in fb.index]
ms["profit"] = [ms.loc[item,"price1"] - ms.loc[item,"Close"] if ms.loc[item,"stock"]==1 else 0 for item in ms.index]
app["profit"] = [app.loc[item,"price1"] - app.loc[item,"Close"] if app.loc[item,"stock"]==1 else 0 for item in app.index]
tsla["profit"] = [tsla.loc[item,"price1"] - tsla.loc[item,"Close"] if tsla.loc[item,"stock"]==1 else 0 for item in tsla.index]
```

```python
# Calculate the cumulative wealth that is received using the cumsum function
fb["wealth"] = fb["profit"].cumsum()
ms["wealth"] = ms["profit"].cumsum()
app["wealth"] = app["profit"].cumsum()
tsla["wealth"] = tsla["profit"].cumsum()
```

```python
# calculate the log of  price1
fb["logprice1"] = np.log(fb["price1"])
ms["logprice1"] = np.log(ms["price1"])
app["logprice1"] = np.log(app["price1"])
tsla["logprice1"] = np.log(tsla["price1"])
```

```python
# calculate the log of close price
fb["logclose"] = np.log(fb["Close"])
ms["logclose"] = np.log(ms["Close"])
app["logclose"] = np.log(app["Close"])
tsla["logclose"] = np.log(tsla["Close"])
```

```python
# calculate the log of the return
fb["logreturn"] = fb["logprice1"] - fb["logclose"]
ms["logreturn"] = ms["logprice1"] - ms["logclose"]
app["logreturn"] = app["logprice1"] - app["logclose"]
tsla["logreturn"] = tsla["logprice1"] - tsla["logclose"]
```

```python
# calculate the mean and sd of fb
mu_fb = fb["logreturn"].mean()
sigma_fb = fb["logreturn"].std(ddof=1)
```

```python
# calculate the mean and sd of ms
mu_ms = ms["logreturn"].mean()
sigma_ms = ms["logreturn"].std(ddof=1)
```

```python
# calculate the mean and sd of app
mu_app = app["logreturn"].mean()
sigma_app = app["logreturn"].std(ddof=1)
```

```python
# calculate the mean and sd of tsla
mu_tsla = tsla["logreturn"].mean()
sigma_tsla = tsla["logreturn"].std(ddof=1)
```

```python
# create a data frame for the probability distribution 
densfb = pd.DataFrame()
densfb["x"] = np.arange(-0.1,0.1,0.001)
densfb["pdf"] = norm.pdf(densfb["x"],mu_fb,sigma_fb)
densfb["cdf"] = norm.cdf(densfb["x"],mu_fb,sigma_fb)
```

```python
# create a data frame for the probability distribution 
densms = pd.DataFrame()
densms["x"] = np.arange(-0.1,0.1,0.001)
densms["pdf"] = norm.pdf(densms["x"],mu_ms,sigma_ms)
densms["cdf"] = norm.cdf(densms["x"],mu_ms,sigma_ms)
```

```python
# create a data frame for the probability distribution 
densapp = pd.DataFrame()
densapp["x"] = np.arange(-0.1,0.1,0.001)
densapp["pdf"] = norm.pdf(densapp["x"],mu_app,sigma_app)
densapp["cdf"] = norm.cdf(densapp["x"],mu_app,sigma_app)
```

```python
# create a data frame for the probability distribution 
denstsla = pd.DataFrame()
denstsla["x"] = np.arange(-0.1,0.1,0.001)
denstsla["pdf"] = norm.pdf(denstsla["x"],mu_tsla,sigma_tsla)
denstsla["cdf"] = norm.cdf(denstsla["x"],mu_tsla,sigma_tsla)
```
